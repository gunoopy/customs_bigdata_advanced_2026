````python
# (코드 실행)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
````

````python
# (코드 실행) 데이터 준비 - 분류
from sklearn.model_selection import train_test_split
df = pd.read_csv('cancer.csv')
X = df.drop('cancer', axis = 1)
y = df['cancer']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

### Bagging (Bootstrap Aggregation)
- 데이터에서 샘플을 여러 번 뽑아(Bootstrap) 학습 결과를 집계(Aggregation)
- Random Forest
    1. 생성할 트리 개수 선택(n_estimators)
    2. 트리 별 부트스트랩 샘플 생성(무작위 복원 추출, max_samples)
    3. 각 샘플 데이터로 트리 학습
        - 각 규칙 생성 시 무작위로 n개의 feature 선택(max_features)
        - 선택된 n개의 feature 중 gini 계수가 가장 많이 감소하는 feature 선택

### Boosting
- 여러 개의 약한 학습기를 순차적으로 연결하여 이전 모델에서 발생한 오류에 대해 높은 가중치를 부여
- GradientBoosting
    - 깊이가 깊지 않은 트리를 여러 개 묶어 사용
    - 이전 트리의 오차를 예측하도록 설계

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier # 분류
from sklearn.ensemble import GradientBoostingRegressorr # 회귀

# XGBOOST
from xgboost import XGBClassifier # 분류
from xgboost import XGBRegressor # 회귀

# LightGBM
from lightgbm import LGBMClassifier # 분류
from lightgbm import LGBMRegressor # 회귀
```
