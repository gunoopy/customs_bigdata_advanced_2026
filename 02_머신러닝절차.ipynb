# scikit-learn을 이용한 머신러닝 절차
1. 데이터 가져오기
    - pandas의 `read_csv()`, `read_excel()` 등의 함수를 이용하여 데이터 불러오기
    - scikit-learn의 datasets의 함수를 이용하여 데이터 불러오기
2. 데이터 전처리(pre-processing)
    - 필수 처리
        - 결측치 처리 - 데이터에 결측치(NaN) 존재 불가
        - 데이터 타입(dtype) 변환 - 반드시 수치형(Numeric) 데이터로 변환
    - 추가 처리 (모델 성능과 연관)
        - 이상치 처리
        - 데이터 보정
3. 특성 공학(Feature Engineering)
    - 모델의 성능을 높이기 위한 과정으로 차원 축소, 스케일링(Scaling), 변환(Transform), 더미(Dummy)와 같은 과정이 존재
    - 처음 머신러닝 모델을 적용하는 경우 생략 후 진행
4. 데이터 분할
    - 데이터를 X와 y로 분리
    - 학습용 데이터(Train)과 평가용 데이터(Test, Valid)로 분할 
        - 주로 7:3 비율로 사용하며 scikit-learn에서는 75:25 비율이 기본 값
        - 보통 `sklearn.model_selection.train_test_split` 함수 사용하여 분리
        - 필수적인 과정은 아니지만 과대 적합(Over-fitting) 방지를 위해 사실상 필수적으로 적용하는 과정
5. 모델 선정
    - 정답(Class, Label, y)의 유형에 따라 모델을 선정
        - 학습시 Label 사용 > 지도학습
            - Label이 범주형 : 분류(Classification) 모델
            - Label이 연속형 : 회귀(Regression) 모델
        - 학습시 Label 미사용 > 비지도학습
            - 군집 모델 등
6. 모델 생성
    ```python
    from sklearn.모델위치 import 모델명
    model = 모델명(파라미터)
    ```
7. 모델 학습
    - `model.fit()` 함수를 이용하여 데이터 학습
    - 지도학습
        - model.fit(X, y)
    - 비지도학습
        - model.fit(X)
8. 모델 평가
     - `model.score()` 함수를 이용하여 모델 성능을 평가
         - model.score(X, y)
     - 모델 유형에 따라 평가 지표가 다름
         - 분류(Classification) 모델 : 정확도(Accuracy)
         - 회귀(Regression) 모델 : $R^2$
9. 모델 예측
     - `model.predict()` 함수를 이용하여 새로운 데이터 예측
         - model.predict(X)
