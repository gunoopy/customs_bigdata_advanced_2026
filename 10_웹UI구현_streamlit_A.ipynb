# 🌸 Iris 머신러닝 + 바이브코딩 웹 시각화 실습

1. **모델 만들기** — `iris.csv`를 불러와 분류 모델을 학습하고 `model.pkl` 한 개로 저장합니다.
2. **바이브코딩으로 웹앱 단계별 구현** — `app.py`(Streamlit)를 4단계로 예쁘게 완성합니다.

| 단계 | 목표 | 결과물 |
|------|------|--------|
| 0단계 | CSV 로드 + 모델 파일 생성 | `model.pkl` |
| 1단계 | 웹 UI 구현 (입력 화면만) | `app.py` |
| 2단계 | 모델 파일 연동 | `app.py` |
| 3단계 | 단일 데이터 예측 | `app.py` |
| 4단계 | CSV 데이터 일괄 예측 | `app.py` |

> **바이브코딩이란?** 코드를 직접 타이핑하지 않고 AI에게 **자연어 프롬프트**로 요청해 코드를 만들어가는 방식입니다. 각 단계마다 **① 문제 → ② 프롬프트 → ③ 결과 코드(app.py)** 순서로 진행합니다.


---
## 0단계 · CSV 로드 & 모델 생성

`iris.csv`를 읽어 모델을 학습하고 `model.pkl` **한 개**로 저장합니다.

````python
import pandas as pd

from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv("iris.csv")
X = df.drop("species", axis=1)
y = df["species"]

model = RandomForestClassifier(random_state=42)
model.fit(X, y)
model.score(X, y)
````

````python
import joblib
joblib.dump(model, "model.pkl")
````

## 🧑‍💻 바이브코딩 시작

각 단계의 코드 셀 맨 위 `%%writefile app.py` 를 실행하면 그 내용이 `app.py`로 저장됩니다. 실제 바이브코딩에서는 함께 적힌 **프롬프트**를 AI에게 주고 결과를 받습니다.

**웹앱 실행** (터미널):
```bash
streamlit run app.py
```

## 1단계 · 웹 UI 구현 (입력 화면만)

### 🎯 문제
사용자가 꽃 측정값 4개를 입력할 수 있는 **예쁜 화면**을 만듭니다. 슬라이더 4개 + 예측 버튼.

### 💬 프롬프트
> Streamlit으로 붓꽃 품종 예측기의 **UI 뼈대**를 예쁘게 만들어줘. 상단에 그라데이션 배경의 제목 배너를 넣고, 꽃받침/꽃잎의 길이·너비 4개를 **2단 컬럼**으로 나눠 슬라이더로 입력받아줘. "예측하기" 버튼은 가로 꽉 차게 하고, 지금은 모델 없이 안내만 띄워줘.

### 🧾 결과 → `app.py`

````python
%%writefile app.py
import streamlit as st

st.set_page_config(page_title="붓꽃 품종 예측기", page_icon="🌸", layout="centered")

st.markdown(
    """
    <div style="text-align:center; padding:1.4rem; border-radius:18px;
         background:linear-gradient(135deg,#a18cd1,#fbc2eb); color:white; margin-bottom:1.2rem;">
        <h1 style="margin:0;">🌸 붓꽃 품종 예측기</h1>
        <p style="margin:.3rem 0 0; opacity:.9;">꽃의 측정값을 입력하면 품종을 예측합니다</p>
    </div>
    """,
    unsafe_allow_html=True,
)

st.subheader("📏 측정값 입력")
c1, c2 = st.columns(2)
with c1:
    sepal_length = st.slider("꽃받침 길이 (cm)", 4.0, 8.0, 5.4, 0.1)
    petal_length = st.slider("꽃잎 길이 (cm)",   1.0, 7.0, 1.3, 0.1)
with c2:
    sepal_width  = st.slider("꽃받침 너비 (cm)", 2.0, 4.5, 3.4, 0.1)
    petal_width  = st.slider("꽃잎 너비 (cm)",   0.1, 2.5, 0.2, 0.1)

if st.button("🔮 예측하기", use_container_width=True):
    st.info("다음 단계에서 모델을 연동합니다! 🙂")
````

✅ `streamlit run app.py` → 그라데이션 배너와 2단 슬라이더, 예측 버튼이 보이면 성공.

## 2단계 · 모델 파일 연동

### 🎯 문제
0단계에서 만든 `model.pkl`을 앱이 불러오도록 합니다. 예측은 아직 안 하고 로딩만 확인.

### 💬 프롬프트
> app.py에 모델 로딩을 추가해줘. joblib으로 model.pkl을 불러오고 `@st.cache_resource`로 캐싱해줘. 잘 불러왔으면 사이드바에 "모델 로드 완료"와 예측 가능한 품종 목록을 예쁘게 보여줘.

### 🧾 결과 → `app.py`

````python
%%writefile app.py
import joblib
import streamlit as st

st.set_page_config(page_title="붓꽃 품종 예측기", page_icon="🌸", layout="centered")

@st.cache_resource
def load_model():
    return joblib.load("model.pkl")

model = load_model()

with st.sidebar:
    st.success("모델 로드 완료 ✅")
    st.caption("예측 가능한 품종")
    for name in model.classes_:
        st.write("🌼", name)

st.markdown(
    """
    <div style="text-align:center; padding:1.4rem; border-radius:18px;
         background:linear-gradient(135deg,#a18cd1,#fbc2eb); color:white; margin-bottom:1.2rem;">
        <h1 style="margin:0;">🌸 붓꽃 품종 예측기</h1>
        <p style="margin:.3rem 0 0; opacity:.9;">꽃의 측정값을 입력하면 품종을 예측합니다</p>
    </div>
    """,
    unsafe_allow_html=True,
)

st.subheader("📏 측정값 입력")
c1, c2 = st.columns(2)
with c1:
    sepal_length = st.slider("꽃받침 길이 (cm)", 4.0, 8.0, 5.4, 0.1)
    petal_length = st.slider("꽃잎 길이 (cm)",   1.0, 7.0, 1.3, 0.1)
with c2:
    sepal_width  = st.slider("꽃받침 너비 (cm)", 2.0, 4.5, 3.4, 0.1)
    petal_width  = st.slider("꽃잎 너비 (cm)",   0.1, 2.5, 0.2, 0.1)

if st.button("🔮 예측하기", use_container_width=True):
    st.info("모델 준비 완료! 다음 단계에서 예측을 연결합니다. 🙂")
````

✅ 사이드바에 "모델 로드 완료 ✅"와 품종 3개가 뜨면 성공. (안 되면 0단계를 실행해 `model.pkl`을 같은 폴더에 만드세요.)

## 3단계 · 단일 데이터 예측

### 🎯 문제
슬라이더로 입력한 값 한 건을 예측하고, **결과 카드 + 신뢰도 + 확률 그래프**로 예쁘게 보여줍니다.

### 💬 프롬프트
> "예측하기" 버튼이 실제로 동작하게 해줘. 슬라이더 값을 데이터프레임으로 만들어 `model.predict`로 품종을 예측하고, 품종 이모지와 이름을 큰 카드로 강조해줘. `model.predict_proba`의 최고 확률을 `st.metric`으로 "신뢰도"로 보여주고, 품종별 확률은 막대그래프로 보여줘.

### 🧾 결과 → `app.py`

````python
%%writefile app.py
import joblib
import pandas as pd
import streamlit as st

st.set_page_config(page_title="붓꽃 품종 예측기", page_icon="🌸", layout="centered")

EMOJI = {"setosa": "🌱", "versicolor": "🌸", "virginica": "🌺"}

@st.cache_resource
def load_model():
    return joblib.load("model.pkl")

model = load_model()

with st.sidebar:
    st.success("모델 로드 완료 ✅")
    st.caption("예측 가능한 품종")
    for name in model.classes_:
        st.write(EMOJI.get(name, "🌼"), name)

st.markdown(
    """
    <div style="text-align:center; padding:1.4rem; border-radius:18px;
         background:linear-gradient(135deg,#a18cd1,#fbc2eb); color:white; margin-bottom:1.2rem;">
        <h1 style="margin:0;">🌸 붓꽃 품종 예측기</h1>
        <p style="margin:.3rem 0 0; opacity:.9;">꽃의 측정값을 입력하면 품종을 예측합니다</p>
    </div>
    """,
    unsafe_allow_html=True,
)

st.subheader("📏 측정값 입력")
c1, c2 = st.columns(2)
with c1:
    sepal_length = st.slider("꽃받침 길이 (cm)", 4.0, 8.0, 5.4, 0.1)
    petal_length = st.slider("꽃잎 길이 (cm)",   1.0, 7.0, 1.3, 0.1)
with c2:
    sepal_width  = st.slider("꽃받침 너비 (cm)", 2.0, 4.5, 3.4, 0.1)
    petal_width  = st.slider("꽃잎 너비 (cm)",   0.1, 2.5, 0.2, 0.1)

if st.button("🔮 예측하기", use_container_width=True):
    X_input = pd.DataFrame([[sepal_length, sepal_width, petal_length, petal_width]],
                           columns=model.feature_names_in_)
    pred = model.predict(X_input)[0]
    proba = pd.Series(model.predict_proba(X_input)[0], index=model.classes_)

    st.markdown(
        f"""
        <div style="text-align:center; padding:1.2rem; border-radius:16px;
             background:#f3f0ff; border:1px solid #e0d7ff; margin:1rem 0;">
            <div style="font-size:3rem;">{EMOJI.get(pred, "🌼")}</div>
            <h2 style="margin:.2rem 0; color:#6b4fbb;">{pred}</h2>
        </div>
        """,
        unsafe_allow_html=True,
    )
    st.metric("신뢰도", f"{proba.max():.1%}")
    st.subheader("품종별 확률")
    st.bar_chart(proba)
````

✅ 슬라이더를 조정하고 "예측하기"를 누르면 품종 카드·신뢰도·확률 그래프가 나옵니다. 꽃잎을 작게 하면 🌱 setosa 확률이 높아집니다.

## 4단계 · CSV 데이터 일괄 예측

### 🎯 문제
CSV를 업로드하면 모든 행을 예측해 **요약 지표 + 표 + 분포 그래프**로 보여주고, 결과 CSV를 다운로드합니다. (`iris.csv`로 바로 테스트 가능)

### 💬 프롬프트
> app.py를 "단일 예측"과 "CSV 일괄 예측" 탭 2개로 나눠줘. CSV 탭에서는 `st.file_uploader`로 파일을 받아 모델 특성 컬럼만 골라 전체를 예측하고, 상단에 총 건수/품종 수를 `st.metric`으로 요약해줘. 원본에 "예측품종"을 붙인 표와 분포 막대그래프를 보여주고, 결과 다운로드 버튼도 넣어줘.

### 🧾 결과 → `app.py` (최종본)

````python
%%writefile app.py
import joblib
import pandas as pd
import streamlit as st

st.set_page_config(page_title="붓꽃 품종 예측기", page_icon="🌸", layout="centered")

EMOJI = {"setosa": "🌱", "versicolor": "🌸", "virginica": "🌺"}

@st.cache_resource
def load_model():
    return joblib.load("model.pkl")

model = load_model()
features = list(model.feature_names_in_)

with st.sidebar:
    st.success("모델 로드 완료 ✅")
    st.caption("예측 가능한 품종")
    for name in model.classes_:
        st.write(EMOJI.get(name, "🌼"), name)

st.markdown(
    """
    <div style="text-align:center; padding:1.4rem; border-radius:18px;
         background:linear-gradient(135deg,#a18cd1,#fbc2eb); color:white; margin-bottom:1.2rem;">
        <h1 style="margin:0;">🌸 붓꽃 품종 예측기</h1>
        <p style="margin:.3rem 0 0; opacity:.9;">단일 예측과 CSV 일괄 예측을 지원합니다</p>
    </div>
    """,
    unsafe_allow_html=True,
)

tab1, tab2 = st.tabs(["🌼 단일 예측", "📄 CSV 일괄 예측"])

# ---------- 단일 예측 ----------
with tab1:
    c1, c2 = st.columns(2)
    with c1:
        sepal_length = st.slider("꽃받침 길이 (cm)", 4.0, 8.0, 5.4, 0.1)
        petal_length = st.slider("꽃잎 길이 (cm)",   1.0, 7.0, 1.3, 0.1)
    with c2:
        sepal_width  = st.slider("꽃받침 너비 (cm)", 2.0, 4.5, 3.4, 0.1)
        petal_width  = st.slider("꽃잎 너비 (cm)",   0.1, 2.5, 0.2, 0.1)

    if st.button("🔮 예측하기", use_container_width=True):
        X_input = pd.DataFrame([[sepal_length, sepal_width, petal_length, petal_width]],
                               columns=features)
        pred = model.predict(X_input)[0]
        proba = pd.Series(model.predict_proba(X_input)[0], index=model.classes_)
        st.markdown(
            f"""
            <div style="text-align:center; padding:1.2rem; border-radius:16px;
                 background:#f3f0ff; border:1px solid #e0d7ff; margin:1rem 0;">
                <div style="font-size:3rem;">{EMOJI.get(pred, "🌼")}</div>
                <h2 style="margin:.2rem 0; color:#6b4fbb;">{pred}</h2>
            </div>
            """,
            unsafe_allow_html=True,
        )
        st.metric("신뢰도", f"{proba.max():.1%}")
        st.bar_chart(proba)

# ---------- CSV 일괄 예측 ----------
with tab2:
    st.caption("필요한 컬럼: " + ", ".join(features))
    file = st.file_uploader("CSV 파일 업로드", type=["csv"])

    if file is not None:
        df = pd.read_csv(file)
        missing = [c for c in features if c not in df.columns]
        if missing:
            st.error("없는 컬럼: " + ", ".join(missing))
        else:
            result = df.copy()
            result["예측품종"] = model.predict(df[features])

            m1, m2 = st.columns(2)
            m1.metric("총 예측 건수", f"{len(result):,}")
            m2.metric("발견된 품종 수", result["예측품종"].nunique())

            st.dataframe(result, use_container_width=True)
            st.subheader("예측 품종 분포")
            st.bar_chart(result["예측품종"].value_counts())

            csv = result.to_csv(index=False).encode("utf-8-sig")
            st.download_button("⬇️ 결과 CSV 다운로드", csv, "predictions.csv",
                               "text/csv", use_container_width=True)
````

✅ "CSV 일괄 예측" 탭에서 `iris.csv`를 업로드하면 요약 지표·결과 표·분포 그래프·다운로드 버튼이 나타납니다.
