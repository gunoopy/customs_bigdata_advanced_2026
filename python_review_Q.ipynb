# 파이썬 리뷰

실무 데이터를 다루기 전, 파이썬의 **기본 문법 · 자료구조 · 표준 라이브러리**를 빠르게 복습하는 노트북입니다.
파이썬은 문법이 간결해 읽기 쉽고, 데이터 처리와 자동화에 널리 쓰입니다.

**사용법**: 설명을 읽고 → 코드 셀을 실행(Shift+Enter)하며 결과를 확인하세요. 각 단원 끝의 `# TODO` 문제는 직접 풀어봅니다.

> pandas 는 별도 노트북에서 다룹니다. 여기서는 외부 패키지 없이 표준 라이브러리만 사용합니다.

**순서**: 변수/자료형 · 연산자 · 문자열 · list · dict · 조건문 · 반복문 · 함수 · 표준 라이브러리(random·datetime)

## 1. 변수와 자료형
값을 이름표에 담아 두는 것이 **변수**이고, 담긴 값의 종류를 **자료형**이라고 합니다.
기본 자료형은 정수(int) · 실수(float) · 문자열(str) · 참거짓(bool) 네 가지입니다.
한 번 만든 변수는 언제든 다시 꺼내 쓰거나 값을 바꿀 수 있습니다.

## 2. 연산자
값을 계산하는 기호입니다. **산술** 연산자(`+ - * / // % **`)로 수를 계산합니다.
`/`는 나누기, `//`는 몫, `%`는 나머지, `**`는 거듭제곱을 뜻합니다.

## 3. 문자열
문자열은 따옴표로 감싼 글자들의 나열입니다.
**f-string**(`f"..."`)으로 변수 값을 문장에 끼워 넣고, **슬라이싱**(`[시작:끝]`)으로 일부만 잘라냅니다.
자주 쓰는 메서드로 `.strip()`(양쪽 공백 제거)과 `.split()`(구분자로 쪼개 리스트로 만들기)이 있습니다.

## 4. list (리스트)
순서가 있는 값들의 묶음입니다. 대괄호 `[]`로 만들고 **인덱스**(0부터 시작)로 각 요소에 접근합니다.
`.append()`로 뒤에 값을 추가하고 `.sort()`로 정렬합니다.
실무 데이터에서 여러 항목을 담는 기본 그릇 역할을 합니다.

## 5. dict (딕셔너리)
**키(key)**로 **값(value)**을 찾는 묶음입니다. 중괄호 `{}` 안에 `키: 값` 쌍으로 만들고 `d[키]`로 값을 꺼냅니다.
없을 수도 있는 키는 `.get(키, 기본값)`으로 안전하게 꺼내고, `.items()`로 키와 값을 함께 순회합니다.
한 건의 데이터(한 행)를 표현할 때 유용합니다.

## 6. 조건문
조건이 참일 때만 특정 코드를 실행합니다.
`if`로 첫 조건을, `elif`로 추가 조건을, `else`로 나머지 경우를 처리합니다.
각 블록은 들여쓰기(보통 공백 4칸)로 구분합니다.

## 7. 반복문
같은 작업을 여러 번 반복할 때 씁니다.
`for 변수 in 리스트:` 형태로 리스트의 요소를 하나씩 꺼내며 실행합니다.
합계나 건수를 집계하는 등 데이터를 훑는 작업에 자주 쓰입니다.

## 8. 함수
자주 쓰는 코드를 `def`로 묶어 이름을 붙인 것입니다.
값을 **인자**로 받아 처리한 뒤 `return`으로 결과를 돌려줍니다.
한 번 정의하면 이름만으로 몇 번이든 재사용할 수 있습니다.

## 9. 표준 라이브러리 (random, datetime)
파이썬에는 설치 없이 바로 쓸 수 있는 표준 라이브러리가 많습니다. `import` 로 불러와 사용합니다.
여기서는 무작위 값을 다루는 **random** 과 날짜를 다루는 **datetime** 을 살펴봅니다.

# 🚀 바이브코딩 연습문제 (실전 종합)

앞서 복습한 **random · list · dict · 조건문 · 반복문 · 함수** 를 모두 활용하는 종합 문제입니다.
각 문제의 **[요구 사항]** 을 읽고, 바로 아래 코드 셀에 직접 구현해보세요.
(각 문제에는 `✅ 정답 예시` 가 함께 제공됩니다. 먼저 스스로 작성한 뒤 비교해보세요.)

### 문제 1. 로또 당첨 시뮬레이터 (일확천금의 꿈)

사용자가 입력한 금액만큼 로또를 자동 구매하고, 랜덤으로 추첨된 당첨 번호와 비교하여 실제 당첨 결과를 확인합니다. 1등부터 5등까지의 당첨 횟수와 총 당첨금을 계산합니다.

**[요구 사항]**
- **구매**: 사용자에게 금액을 입력받아 1,000원당 1장씩 로또 번호(1~45 중 6개)를 자동 생성
- **추첨**: 이번 회차의 당첨 번호 6개와 보너스 번호 1개를 랜덤 생성 (중복 불가)
- **판정**:
  - 1등: 6개 일치 / 2등: 5개 + 보너스 일치 / 3등: 5개 일치 / 4등: 4개 일치 / 5등: 3개 일치
- **당첨금**:
  - 1등: 20억 / 2등: 5천만 / 3등: 150만 / 4등: 5만 / 5등: 5천
- **결과**: 등수별 당첨 횟수와 총 당첨 금액을 합산하여 수익률과 함께 출력

### 문제 2. 가위바위보 3판 2선승 챔피언

사용자와 컴퓨터가 가위바위보 대결을 펼칩니다. 먼저 2승을 거두는 쪽이 최종 승리하며, 매 판 결과를 기록하고 마지막에 전적과 승률을 출력합니다.

**[요구 사항]**
- **입력**: 매 판마다 사용자에게 `가위`, `바위`, `보` 중 하나를 입력받음 (잘못 입력 시 다시 입력)
- **컴퓨터**: 매 판마다 셋 중 하나를 랜덤 선택
- **판정**:
  - 승리: 가위>보, 바위>가위, 보>바위 / 무승부: 같은 것 / 나머지는 패배
  - 무승부는 승패에 포함하지 않고 다시 진행
- **종료 조건**: 사용자 또는 컴퓨터가 먼저 2승에 도달하면 종료
- **결과**: 최종 승자, 총 진행 판수, 사용자 전적(승-무-패), 승률(%)을 출력

### 문제 3. 숫자 야구 게임 (Number Baseball)

컴퓨터가 서로 다른 3자리 숫자를 정하면, 사용자는 이를 맞히기 위해 숫자를 추측합니다. 자리와 숫자가 모두 맞으면 스트라이크, 숫자는 맞지만 자리가 틀리면 볼로 힌트를 제공합니다.

**[요구 사항]**
- **정답 생성**: 1~9 사이의 서로 다른 숫자 3개를 랜덤으로 선택 (0 미포함, 중복 불가)
- **입력**: 사용자에게 3자리 숫자를 입력받음 (서로 다른 숫자 3개인지 검증)
- **판정**:
  - 스트라이크(S): 숫자와 자리가 모두 일치
  - 볼(B): 숫자는 있지만 자리가 다름
  - 아웃(OUT): 하나도 일치하지 않음
- **종료 조건**: 3 스트라이크가 되면 정답 처리하고 종료
- **결과**: 매 시도마다 `nS mB` 형태로 힌트 출력, 정답 시 총 시도 횟수 출력
