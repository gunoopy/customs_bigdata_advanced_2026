### 문제 1. 로또 당첨 시뮬레이터 (일확천금의 꿈)

사용자가 입력한 금액만큼 로또를 자동 구매하고, 랜덤으로 추첨된 당첨 번호와 비교하여 실제 당첨 결과를 확인합니다. 1등부터 5등까지의 당첨 횟수와 총 당첨금을 계산합니다.

**[요구 사항]**
- **구매**: 사용자에게 금액을 입력받아 1,000원당 1장씩 로또 번호(1~45 중 6개)를 자동 생성
- **추첨**: 이번 회차의 당첨 번호 6개와 보너스 번호 1개를 랜덤 생성 (중복 불가)
- **판정**:
  - 1등: 6개 일치 / 2등: 5개 + 보너스 일치 / 3등: 5개 일치 / 4등: 4개 일치 / 5등: 3개 일치
- **당첨금**:
  - 1등: 20억 / 2등: 5천만 / 3등: 150만 / 4등: 5만 / 5등: 5천
- **결과**: 등수별 당첨 횟수와 총 당첨 금액을 합산하여 수익률과 함께 출력

```python
# ✅ 정답 예시
import random

def buy_lotto(count):
    # 1,000원당 1장씩, 1~45 중 6개를 뽑아 count장 생성
    return [sorted(random.sample(range(1, 46), 6)) for _ in range(count)]

def draw_winning():
    # 당첨 번호 6개 + 보너스 1개 추첨 (중복 없음)
    picked = random.sample(range(1, 46), 7)
    return set(picked[:6]), picked[6]

def check_rank(ticket, win_set, bonus):
    # 티켓 1장의 등수 판정 (0이면 낙첨)
    match = len(set(ticket) & win_set)
    if match == 6:
        return 1
    if match == 5 and bonus in ticket:
        return 2
    if match == 5:
        return 3
    if match == 4:
        return 4
    if match == 3:
        return 5
    return 0

def simulate_lotto():
    money = int(input("구매 금액을 입력하세요 (원): "))
    count = money // 1000

    tickets = buy_lotto(count)
    win_set, bonus = draw_winning()

    prize = {1: 2_000_000_000, 2: 50_000_000, 3: 1_500_000, 4: 50_000, 5: 5_000}
    result = {rank: 0 for rank in range(1, 6)}

    for t in tickets:
        rank = check_rank(t, win_set, bonus)
        if rank:
            result[rank] += 1

    print(f"\n구매 매수 : {count:,}장")
    print(f"당첨 번호 : {sorted(win_set)} + 보너스 {bonus}\n")

    total = 0
    for rank in range(1, 6):
        amount = result[rank] * prize[rank]
        total += amount
        print(f"{rank}등 : {result[rank]:>6,}회  ->  {amount:>15,}원")

    rate = (total - money) / money * 100 if money else 0
    print(f"\n투자 금액 : {money:,}원")
    print(f"총 당첨금 : {total:,}원")
    print(f"수익률    : {rate:,.2f}%")

simulate_lotto()
```

### 문제 2. 가위바위보 3판 2선승 챔피언

사용자와 컴퓨터가 가위바위보 대결을 펼칩니다. 먼저 2승을 거두는 쪽이 최종 승리하며, 매 판 결과를 기록하고 마지막에 전적과 승률을 출력합니다.

**[요구 사항]**
- **입력**: 매 판마다 사용자에게 `가위`, `바위`, `보` 중 하나를 입력받음 (잘못 입력 시 다시 입력)
- **컴퓨터**: 매 판마다 셋 중 하나를 랜덤 선택
- **판정**:
  - 승리: 가위>보, 바위>가위, 보>바위 / 무승부: 같은 것 / 나머지는 패배
  - 무승부는 승패에 포함하지 않고 다시 진행
- **종료 조건**: 사용자 또는 컴퓨터가 먼저 2승에 도달하면 종료
- **결과**: 최종 승자, 총 진행 판수, 사용자 전적(승-무-패), 승률(%)을 출력

```python
# ✅ 정답 예시
import random

HANDS = ["가위", "바위", "보"]
# 키가 값을 이긴다
WINS = {"가위": "보", "바위": "가위", "보": "바위"}

def get_user_hand():
    while True:
        hand = input("가위 / 바위 / 보 중 입력: ").strip()
        if hand in HANDS:
            return hand
        print("  ⚠️  '가위', '바위', '보' 중에서 입력하세요.")

def play_rps():
    user_win, com_win, draw = 0, 0, 0
    rounds = 0

    while user_win < 2 and com_win < 2:
        rounds += 1
        user = get_user_hand()
        com = random.choice(HANDS)
        print(f"[{rounds}판] 나: {user}  vs  컴퓨터: {com}", end="  ->  ")

        if user == com:
            draw += 1
            print("무승부 🤝")
        elif WINS[user] == com:
            user_win += 1
            print("승리 🎉")
        else:
            com_win += 1
            print("패배 😢")

    winner = "사용자" if user_win == 2 else "컴퓨터"
    decided = user_win + com_win
    rate = user_win / decided * 100 if decided else 0

    print("\n" + "=" * 30)
    print(f"🏆 최종 승자 : {winner}")
    print(f"총 진행 판수 : {rounds}판")
    print(f"사용자 전적 : {user_win}승 {draw}무 {com_win}패")
    print(f"승률(무승부 제외) : {rate:.1f}%")

play_rps()
```

### 문제 3. 숫자 야구 게임 (Number Baseball)

컴퓨터가 서로 다른 3자리 숫자를 정하면, 사용자는 이를 맞히기 위해 숫자를 추측합니다. 자리와 숫자가 모두 맞으면 스트라이크, 숫자는 맞지만 자리가 틀리면 볼로 힌트를 제공합니다.

**[요구 사항]**
- **정답 생성**: 1~9 사이의 서로 다른 숫자 3개를 랜덤으로 선택 (0 미포함, 중복 불가)
- **입력**: 사용자에게 3자리 숫자를 입력받음 (서로 다른 숫자 3개인지 검증)
- **판정**:
  - 스트라이크(S): 숫자와 자리가 모두 일치
  - 볼(B): 숫자는 있지만 자리가 다름
  - 아웃(OUT): 하나도 일치하지 않음
- **종료 조건**: 3 스트라이크가 되면 정답 처리하고 종료
- **결과**: 매 시도마다 `nS mB` 형태로 힌트 출력, 정답 시 총 시도 횟수 출력

```python
# ✅ 정답 예시
import random

def make_answer():
    # 1~9 중 서로 다른 숫자 3개
    return random.sample(range(1, 10), 3)

def get_guess():
    while True:
        raw = input("3자리 숫자를 입력하세요: ").strip()
        if (len(raw) == 3 and raw.isdigit()
                and "0" not in raw and len(set(raw)) == 3):
            return [int(c) for c in raw]
        print("  ⚠️  1~9 사이 서로 다른 숫자 3개를 입력하세요.")

def judge(answer, guess):
    strike = sum(a == g for a, g in zip(answer, guess))
    ball = sum(g in answer for g in guess) - strike
    return strike, ball

def play_baseball():
    answer = make_answer()
    tries = 0
    print("⚾ 숫자 야구 게임 시작! (1~9, 서로 다른 3자리)\n")

    while True:
        guess = get_guess()
        tries += 1
        strike, ball = judge(answer, guess)

        if strike == 3:
            print(f"\n🎉 정답! ({''.join(map(str, answer))})")
            print(f"총 시도 횟수 : {tries}번")
            break
        elif strike == 0 and ball == 0:
            print("  결과: OUT 💥")
        else:
            print(f"  결과: {strike}S {ball}B")

play_baseball()
```
